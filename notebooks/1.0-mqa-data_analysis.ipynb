```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys

sys.path.append('../../src/pre_processing.py')
from pre_processing import BTCUSD_Features
```

```python
BTCUSD = BTCUSD_Features()
BTCUSD_df = BTCUSD.get_full_df()
```

```python
fig, axs = plt.subplots(nrows=5, ncols=2, figsize=(12, 12))
plt.subplots_adjust(top=1.25, bottom=0.2, hspace=0.4, wspace=0.2)

sns.scatterplot(x='Open', y='High', data=BTCUSD_df, ax=axs[0,0])
sns.scatterplot(x='Close', y='High', data=BTCUSD_df, ax=axs[0,1])
sns.scatterplot(x='High', y='High', data=BTCUSD_df, ax=axs[1,0])
sns.scatterplot(x='Low', y='High', data=BTCUSD_df, ax=axs[1,1])
sns.scatterplot(x='Volume', y='High', data=BTCUSD_df, ax=axs[2,0])
sns.scatterplot(x='BTCPJPY', y='High', data=BTCUSD_df, ax=axs[2,1])
sns.scatterplot(x='BTCEUR', y='High', data=BTCUSD_df, ax=axs[3,0])
sns.scatterplot(x='ETHUSD', y='High', data=BTCUSD_df, ax=axs[3,1])
sns.scatterplot(x='XRPUSD', y='High', data=BTCUSD_df, ax=axs[4,0])
sns.scatterplot(x='DOGEUSD', y='High', data=BTCUSD_df, ax=axs[4,1])

plt.show()
```

```python
# Correlation heatmap
corr = BTCUSD_df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
ma_day = [30, 100, 200]
features = ['High', 'Low', 'BTCPJPY', 'BTCEUR', 'ETHUSD', 'XRPUSD', 'DOGEUSD']
moving_avgs = pd.DataFrame(columns=[f"{ma} Candle MA for {column}" for ma in ma_day for column in features])
# print(BTCUSD_df.to_string())
for ma in ma_day:
    for column in features:
        column_name = f"{ma} Candle MA for {column}"
        moving_avgs[column_name] = BTCUSD_df[column].rolling(ma).mean()
print(moving_avgs.to_string())
# moving_avgs.groupby("Date").hist(figsize=(18,12))
# ma_df = BTCUSD_Features.get_moving_avgs_df(BTCUSD_df, 30, 100, 200)
```

```python
for feature in BTCUSD_df.columns[:-1]:
    ax = sns.boxplot(x=BTCUSD_df['High'], y=BTCUSD_df[feature])
    ax.set_ylabel('Boxplot Across Features')
```

```python
fig, axs = plt.subplots(nrows=5, ncols=2, figsize=(12, 20))
for i, feature in enumerate(BTCUSD_df.columns[:-1]):
    axs[i//2, i%2].plot(BTCUSD_df.index, BTCUSD_df[feature])
    axs[i//2, i%2].set_title(feature)
    axs[i//2, i%2].set_xlabel('Date')
    axs[i//2, i%2].set_ylabel(feature)
plt.tight_layout()
plt.show()
```
